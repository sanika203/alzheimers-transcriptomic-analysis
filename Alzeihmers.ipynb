```python
import pandas as pd
#reading the data file
df = pd.read_csv(
    "/content/GSE44770_originalDataFile_PFC.txt.gz",
    sep="\t",
    compression="gzip",
    comment="!"
)
df.head()




```

```python
df.columns[:10]
```

```python
raw = pd.read_csv(
    "/content/GSE44770_originalDataFile_PFC.txt.gz",
    sep="\t",
    compression="gzip",
    header=None
)

raw.head(15)
```

```python
expr_cols = df.columns[3:]
len(expr_cols)
#RNA expression levels start from column 3
```

```python
half = len(expr_cols) // 2

control_cols = expr_cols[:half]
ad_cols = expr_cols[half:]
df["Control_mean"] = df[control_cols].mean(axis=1)
df["AD_mean"] = df[ad_cols].mean(axis=1)
import numpy as np

df["log2FC"] = np.log2((df["AD_mean"] + 1e-6) / (df["Control_mean"] + 1e-6))
df[["log2FC"]].head()
```

```python
# Sorted by log2 fold change
top_up = df.sort_values("log2FC", ascending=False).head(10)
top_down = df.sort_values("log2FC", ascending=True).head(10)

top_up[["gene", "log2FC"]], top_down[["gene", "log2FC"]]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(range(len(df)), df["log2FC"], alpha=0.5)
plt.axhline(0, color="black", linestyle="--")

plt.xlabel("Genes")
plt.ylabel("log2 Fold Change (AD vs Control)")
plt.title("Differential Expression in Alzheimerâ€™s Disease (PFC)")

plt.show()
```

```python
df_out = df[["gene", "log2FC"]].sort_values("log2FC", ascending=False)
df_out.to_csv("AD_PFC_log2FC_results.csv", index=False)
```
